```python
# import librares
from os import listdir
from os.path import isfile, join
import os
import gpxpy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
# Initiate a main Dataframe
main_df = pd.DataFrame(columns = ['day','lon','lat'])
```

```python
# Define counter for the days
counter = 1

# Read gpx files from directory
for gpx_file in os.listdir('../data/scheduled_trip/')[1:]:
    gpx_open = open('../data/scheduled_trip/'+gpx_file, 'r')
    
    # Parse each file
    gpx = gpxpy.parse(gpx_open)
    
    # Create a dictionary 
    temp_dict = defaultdict(dict)
    
    # Add long lat to the dictionary
    for track in gpx.tracks:
        for segment in track.segments:
            temp_dict['lon'] = [point.longitude for point in segment.points]
            temp_dict['lat'] = [point.latitude for point in segment.points]
    
    # Create a temporary dataframe
    temp_df = pd.DataFrame(temp_dict)
    
    # Add day number into the df
    temp_df['day'] = pd.Series(counter, index=temp_df.index)
    
    # Append to the main dataframe
    main_df = pd.concat([main_df,temp_df])
    
    # Increment counter for each day
    counter +=1
```

```python
# Export to csv
main_df.to_csv('../data/scheduled_trip.csv', index=False )
```
